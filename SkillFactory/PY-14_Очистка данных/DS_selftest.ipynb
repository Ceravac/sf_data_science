```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
Data_Parsing = pd.read_excel('data/Data_TSUM.xlsx', sheet_name='Data_Parsing')
Data_Company = pd.read_excel('data/Data_TSUM.xlsx', sheet_name='Data_Company')


def cleaner_df(colum): #функция замены символов 
    if type(colum) is not int:
        for simbol in ['_', '-', '~', '\\', '/']:
            colum = colum.replace(simbol, '')
    return colum


for col in Data_Parsing.columns:#применяя функцию к таблице парсинга дабы убрать лишние символы
    Data_Parsing[col] = Data_Parsing[col].apply(cleaner_df)
    
    
```

```python
display(Data_Parsing)
```

```python
# обьединение таблиц
Data_Parsing_coop = Data_Parsing.merge(Data_Company, how='outer', right_on='item_id', left_on='producer_id').drop(labels = ['Category_y','brand_y'], axis=1)
Data_Parsing_coop['difference'] = (Data_Parsing_coop['price'] / Data_Parsing_coop['current price'] *100).round(2)
display(Data_Parsing_coop)
```
